```python
users = ['sushant', 'tanmay', 'pranay', 'ankush', 'kiran']

print('*** Welcome to Shipping Panel ***')

user = input('Enter your username: ').lower().strip()

if user in users:
    print(f'Hello {user.capitalize()}, Welcome back to your account!\n')
    print('Current Shipping charges are below:')
    print('Shipping Orders 0 to 100: \tRs.100/item')
    print('Shipping Orders 100 to 150: \tRs.80/item')
    print('Shipping Orders 150 to 200: \tRs.50/item')
    print('Shipping Orders more than 200: \tFree Delivery')

    ship_items = int(input('How many items would you like to ship: '))

    if ship_items < 100:
        price = ship_items * 100
    elif ship_items < 150:
        price = ship_items * 80
    elif ship_items < 200:
        price = ship_items * 50
    else:
        price = 0

    print(f'For {ship_items} items, the shipping charges are Rs. {price}.')

    confirmation = input('Would you like to place this order? (Yes/No): ').strip().lower()
    if confirmation == 'yes':
        print(f'Thank you! Your order for {ship_items} items has been placed successfully!')
    else:
        print('No order has been placed.')

else:
    print('Sorry! You are not registered in our system.\nPlease create an account first.')
```
