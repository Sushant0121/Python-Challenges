CHALLENGE_2 :  You are responsible for writing a program that will simulate registering to vote. If a user is 18 or older, your program will present them with a list of potential political parties to register for. Upon choosing a party, your program will confirm that the user has registered and print a specific message depending on the party joined.

```python
# Voting App

print('Loksabha Election 2025')

# user info

userName = input('Enter your name: ')
userAge = int(input('Enter your age(as per voterId):'))

political_parties = ['BJP','SP', 'CONG', 'BSP', 'APP', 'OTHER', 'NOTA']

# condition

if userAge >= 18:
  # do something
  print('You are Eligible to Vote! Please select a party to vote')
  for party in political_parties:
    print(f'-\t {party}')

  userParty = input('Which party would you like to vote? Please enter the name:').strip().upper()

  if userParty in political_parties:
    print('Thank you for voting! Jai Hind')
  else:
    print('This is INVALID.!')


else:
  print(f'You are not eligible to Vote! See you next time after {18-userAge}years!')
```
