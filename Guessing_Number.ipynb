```python
import random


# Welcome Message
print('*** Welcome to Number Gussing Game! ***')


# Get user name
name = input('Enter your Name: ')


# Print instruction
print(f'Hello {name}, I am thinking a number which is between 1 to 20.\nYou have 5 Guesses to guess the number!')

number = random.randint(1,20)


for guess_count in range(5):
  print(f'Round {guess_count + 1}')
  guess = int(input('Take a Guess: '))


  if guess > number:
    print('Your Guess is too HIGH! Try Again')
  elif guess < number:
    print('Your Guess is too LOW! Try something Higher!')
  else:
    break


if guess == number:
  print(f'Hurray! {name}. You guessed my number in {guess_count + 1} guesses.')
else:
  print(f'GAME OVER! You Loose! \nMy number was {number}')
```
